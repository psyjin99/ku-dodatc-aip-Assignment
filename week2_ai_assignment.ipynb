## [실습 과제 1/3]
1. 1부터 50까지의 숫자 중 짝수만 출력하는 `for` 문 또는 `while` 문을 작성해보세요.


```python
even_numbers = []
for a in range(1, 51) :
    if a % 2 == 0 :
        even_numbers += [a]

print(f"1부터 50까지의 짝수는 : {even_numbers}") 
```

2. 두 개의 숫자를 입력받아 곱한 값을 반환하는 `multiply(a, b)` 함수를 만들고 테스트해보세요.

```python
# a, b 값을 함수 밖에서 받고, 함수는 계산만 하기  
def multiply(a, b) :    
    return a*b


a = float(input("첫번째 계산 num1을 입력하세요 : "))
b = float(input("첫번째 계산 num2를 입력하세요 : "))
ab = multiply(a, b)
print(f"첫번째 계산 결과는 {a} X {b} = {ab} 입니다.")



# a, b 값을 함수에서 직접 받도록 정의하고 그 값도 반환하도록 하기 
def multiply_input() :    
    a = float(input("두번째 계산 num1을 입력하세요 : "))
    b = float(input("두번째 계산 num2를 입력하세요 : "))
    return a, b, a*b


num1, num2, result = multiply_input()
print(f"두번째 계산 결과는 {num1} X {num2} = {result} 입니다.")


## 입력은 외부에서 계산은 함수를 이용해서 하는 것이 함수를 재사용하기 좋을 것으 판단 ##
```

## [실습 과제 3/3]
3. `Robot` 클래스에 배터리를 100%로 채우는 `charge()` 메서드를 추가하고 작동을 확인해보세요.

```python
# 로봇 클래스 정의하기 
class Robot:
    
    def __init__(self, name, model, vendor) :
        self.vendor = vendor
        self.name = name
        self.model = model
        self.battery = 100
        
    def greet(self) :
        print(f"안녕하세요! 저는 {self.vendor}에서 만든 {self.model} 모델 {self.name} 입니다.")
        
    def work(self) :
        if self.battery > 20:
            print(f"{self.name}가 작업을 수행합니다. 배터리 잔량 {self.battery} %. 삐빅.")
            self.battery -= 20
        else:
            print("배터리가 부족합니다. 충전해주세요.")

    # 배터리를 충전하는 charge() 매서드 추가하기
    def charge(self) :
        while self.battery < 100:
            print(f"{self.name} 배터리 충전 중. 현재 {self.battery} %. 삐빅.")
            self.battery += 20
            
        else:
            print("배터리 충전이 완료되었습니다.")


# 로봇 객체 생성
robot = Robot(name="Alpha_Go", model="Alpha_atlas", vendor="Aero_Dynamics")


# 로봇의 작동과 충전 테스트
robot.greet()
robot.work()
robot.work()
robot.work()
robot.work()
robot.work()
robot.work()
robot.charge()
robot.charge()
```
